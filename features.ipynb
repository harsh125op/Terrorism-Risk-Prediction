# let's start with feature engenerring

```python
import pandas as pd
```

```python
#Load the dataset
df = pd.read_csv("dataset\india_terrorism_database_with_summary.csv")
```

```python
#convert to data time format
df.columns
```

```python
# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'], errors = 'coerce')
```

```python
# Extract Year and Month from Date
df['Year'] = df['Date'].dt.year
df['Months'] = df['Date'].dt.month
```

```python
df.head()
```

```python
# Define a function to classify the risk category
def classify_risk(cas):
    if cas == 0:
        return "No Risk"
    elif cas <= 5:
        return "Low Risk"
    elif cas <= 20:
        return "Medium Risk"
    else:
        return "High Risk"
```

```python
# add the risk column and add data using the classify risk function
df['Risk'] = df['Casualties'].apply(classify_risk)
```

```python
df.head()
```

```python
#save the csv
df.to_csv(r'dataset\updated_file_with_risk_level.csv', index=False)
```
